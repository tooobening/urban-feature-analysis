```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
# using seaborn, distribution plot, to decide which scheme is suitable for this dataset
import seaborn as sns
#https://github.com/pysal/mapclassify
```

```python
data = gpd.read_file('./data/shp/Sea_cbg_data_final.shp')
```

```python
data
```

```python
sns.distplot(data['D2B_E8MIXA'])
```

```python
### Scheme must be in the set: dict_keys(['boxplot', 'equalinterval', 'fisherjenks', 'fisherjenkssampled', 'headtailbreaks', 'jenkscaspall', 'jenkscaspallforced', 'jenkscaspallsampled', 
#'maxp', 'maximumbreaks', 'naturalbreaks', 'quantiles', 'percentiles', 'stdmean', 'userdefined'])
# https://geopandas.org/reference.html , when plotting legend, Keyword arguments to pass to matplotlib.pyplot.legend() or matplotlib.pyplot.colorbar(). Additional accepted keywords when scheme is specified:
fig, ax = plt.subplots(figsize = (40,30))
data.plot(zorder=1,column='D2B_E8MIXA', cmap='RdPu',edgecolor='black',linewidth=0.5,legend="True",ax=ax, scheme='quantiles',
         markersize=1,legend_kwds={'loc': 'upper right','fontsize':26}) 
plt.title('Mean Employment Mix',fontsize=72) #Higher values correlate with more walk trips
plt.axis('off')
plt.savefig('sea_employ_quantiles.png',dpi=300)
```

```python
sns.distplot(data['dis_ospace'])
```

```python
#Scheme must be in the set: dict_keys(['boxplot', 'equalinterval', 'fisherjenks', 'fisherjenkssampled', 'headtailbreaks', 'jenkscaspall', 'jenkscaspallforced', 'jenkscaspallsampled', 
#'maxp', 'maximumbreaks', 'naturalbreaks', 'quantiles', 'percentiles', 'stdmean', 'userdefined'])
fig, ax = plt.subplots(figsize = (40,30))
data.plot(zorder=1,column='dis_ospace', cmap='YlGn',edgecolor='black',linewidth=0.5,legend="True",ax=ax, scheme='fisherjenks',
         legend_kwds={'loc': 'upper right','fontsize':20}) #edgecolor='black',
plt.title('Mean Distance to \n Nearest Open Space (m)',fontsize=72)
# bg.plot(zorder=2,facecolor="none",edgecolor='black',ax=ax, alpha =1) # no-fill
plt.axis('off')
plt.savefig('sea_dis_ospce_fisherjenks.png',dpi=300)
```
